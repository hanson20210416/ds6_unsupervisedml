### Portfolio Assignment: Identifying Biomarkers for Pre-Eclampsia Using LC-MS Data

Pre-eclampsia is a pregnancy complication characterized by high blood pressure, higher protein abundance in urine and signs of damage to other organs, often the liver or kidneys. The condition can lead to serious, even life-threatening complications for both the mother and baby if not properly managed. Pre-eclampsia is often detected at a late stage. Currently, the earliest diagnosis typically occurs around 20 weeks of pregnancy. Treatment options are limited, often involving early delivery of the baby or administering blood pressure-lowering medications.  Research has shown that taking a low dose of calcium and aspirin before 16 weeks of pregnancy can be effective, but administering this to healthy mothers is not ideal, since there could be side effects of the medicin. Therefore, there is an urgent need for earlier detection to enable more targeted and manageable treatments.

Liquid chromatography coupled with mass spectrometry (LC-MS) has emerged as a highly effective technique for the sensitive detection, quantification, and identification of proteins and peptides in biological fluids, such as serum (Govorukhina, 2006). In the field of proteomics, LC-MS is widely used for the discovery of disease biomarkers

### In the next I am going to apply Decisiontree to find the best feature to split the groups(2 and 4 with others)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

df_clean['group_key_binary'] = df_clean['group_key'].apply(
    lambda x: 'Group2_4' if x in ['group2', 'group4'] else 'Other'
)
X_subset = df_clean.iloc[:, :-2].values  
y_subset = df_clean['group_key_binary']

scaler = StandardScaler()
X_subset = scaler.fit_transform(X_subset)
X_train, X_test, y_train, y_test = train_test_split(X_subset, y_subset, test_size=0.2, random_state=42)
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
importances = clf.feature_importances_
important_features = np.argsort(importances)[::-1]
feature_names = df_clean.columns[:-2] 
print("Top Important Features (Protein Names):")
for i in important_features:
    if importances[i] > 0:
        protein_name = feature_names[i]
        print(f"{protein_name}: Importance {importances[i]:.4f}")
```
